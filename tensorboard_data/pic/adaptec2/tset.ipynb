```python
import matplotlib.pyplot as plt
import pandas as pd

def tensorboard_smoothing(x,smooth=0.99):
    x = x.copy()
    weight = smooth
    for i in range(1,len(x)):
        x[i] = (x[i-1] * weight + x[i]) / (weight + 1)
        weight = (weight + 1) * smooth
    return x
```

```python
fig, ax1 = plt.subplots(1, 1)    # a figure with a 1x1 grid of Axes
#设置上方和右方无框
ax1.spines['top'].set_visible(False)                   # 不显示图表框的上边框
ax1.spines['right'].set_visible(False)  
seed_1 = pd.read_csv("adaptec2-tag-_result_reward.csv")
# seed_3 = pd.read_csv("run-2024-38-192027-tag-_result_reward.csv")
# seed_2 = pd.read_csv("run-2024-38-193415-tag-_result_reward.csv")
# seed_0 = pd.read_csv("run-2024-38-191757-tag-_result_reward.csv")
import numpy as np
#设置折线颜色，折线标签
#使用平滑处理
seed_0 = pd.read_csv("adaptec2-_result_reward.csv")
seed_1['Value'] = np.exp(7-seed_1['Value']/2)

#ax1.plot(wl['Step']//1000, tensorboard_smoothing(wl['Value'], smooth=0.1), color="blue")
# ax1.plot(seed_0['Step']//1000, tensorboard_smoothing(seed_0['Value'], smooth=0), color="orange",label='no cluster')
ax1.plot(seed_1['Step']//1000, tensorboard_smoothing(seed_1['Value'], smooth=0), color="red",label="adaptec2")
# ax1.plot(seed_0['Step']//1000, tensorboard_smoothing(seed_0['Value'], smooth=0), color="blue",label='seed=0')
# ax1.plot(seed_2['Step']//1000, tensorboard_smoothing(seed_2['Value'], smooth=0), color="green",label="seed=10")
# ax1.plot(seed_3['Step']//1000, tensorboard_smoothing(seed_3['Value'], smooth=0), color="black",label="seed=100")
#ax1.plot(seed_2['Step']//1000, tensorboard_smoothing(seed_2['Value'], smooth=0), color="lightcoral",label="adaptec2")

#不使用平滑处理
# ax1.plot(len_mean['Step'], len_mean['Value'], color="red",label='all_data')
fonts = {'family':'Times New Roman','size':14}
#s设置标签位置，lower upper left right，上下和左右组合
plt.legend(loc = 'lower right',prop=fonts)
ax1.set_xlabel("Step(K)",fontdict=fonts)
#ax1.set_ylabel("Wirelength",fontdict=fonts)
ax1.set_ylabel("Reward",fontdict=fonts)

plt.show()
#保存图片，也可以是其他格式，如pdff
fig.savefig(fname='./adaptec2_noEA_reward'+'.png', format='png')
```
